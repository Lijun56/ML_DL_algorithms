# Logistic Regression
## what it is
* step1(regression step)similar to Linear Regression, we find probability as output for certain x: $\widehat{p} = f(x)$ ( similar to $\widehat{y} = f(x)$ in Lin_reg)
* step2(classifier step)
  * $\hat{y} = 
    \begin{cases} 
    1, & \hat{p} \geq 0.5 \\ 
    0, & \hat{p} < 0.5 
    \end{cases}$
* above all, logistic regression usually can only be used for binary classification
  * but multi-classification question can be transformed to binary classfication questions and then use logisticReg method to solve


## how it works
* sigmoid function $\sigma()$
  * $\hat{p} = \sigma(\theta^T \cdot x_b)$
  * $\sigma(t) = \frac{1}{1 + e^{-t}}$
  * ![sigmoid func graph](assets/logisticReg/sigmoidFunc.png)

```python
#sigmoid function
import numpy as np
import matplotlib.pyplot as plt
x = np.linspace(-10, 10, 500)
def sigma(x):
    return 1/(1+np.exp(-x))
plt.title("sigmoid function")
plt.plot(x, sigma(x))

#cost function
# Define the range and the functions
x = np.linspace(0.01, 0.99, 100)  # Avoid 0 and 1 to prevent log issues
y1 = -np.log(x)       # y = -log(x)
y2 = -np.log(1 - x)   # y = -log(1 - x)

# Plot the functions
plt.figure(figsize=(8, 6))
plt.plot(x, y1, label=r"$y=-\log(x)$", color="blue")
plt.plot(x, y2, label=r"$y=-\log(1-x)$", color="red")
plt.scatter([1], [0], color="black", label="(1, 0)")

# Add labels and grid
plt.xlabel("x")
plt.ylabel("y")
plt.title("cost functions")
plt.axvline(1, color="gray", linestyle="--", linewidth=0.5)
plt.legend()
plt.grid(True)

# Set limits to zoom into [0, 1] for x-axis
plt.xlim(0, 2.5)
plt.ylim(0, 5)

plt.show()
```
