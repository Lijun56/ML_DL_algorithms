# Gradient Descents

* not a ml algorithm
* a search algorithm to find the coef that optimzie a function
* used in minizing loss function
* not all function gaurantee to have a pole(direvative = 0), for function that is not unform:
  * we random select point from function multiple times
* how:
  * $ \frac{dJ}{d\theta}$ each time $- (-\eta \frac{dJ}{d\theta})$ until find polar point
  * $\eta$ is learning coef here
  * compare abs of distance from last point to determine the polar point, set a threshold epsilon


##env

```python
import numpy as np
import matplotlib.pyplot as plt
```

## simple GD

```python
x = np.linspace(-1, 6, 141)
y = (x-2.5)**2 - 1
plt.plot(x, y)
```

```python
def dX(x):
  return 2*(x -2.5)

def FindX(x):
  return (x - 2.5)**2 - 1

def Gradient_Descent(initial_theta, eta, epsilon = 1e-8 ):
  theta_history = [initial_theta]
  theta = initial_theta
  while True:
    gradient = dX(theta)
    last_theta = theta
    theta = theta - eta * gradient
    theta_history.append(theta)
    if(abs(FindX(theta) - FindX(last_theta)) < epsilon):
      break;
  return theta_history


def plot_theta_history():
  plt.plot(x, FindX(x))
  plt.plot(np.array(theta_history), FindX(np.array(theta_history)), color = 'r', marker = '+')

```

```python
# test 1
theta = 0
eta = 0.1
theta_history=[]
theta_history = Gradient_Descent(theta, eta)
plot_theta_history()
```
